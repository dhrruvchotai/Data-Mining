```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#age - age in years

#sex - (1 = male; 0 = female)

#cp - chest pain type

#trestbps - resting blood pressure (in mm Hg on admission to the hospital)

#chol - serum cholestoral in mg/dl no ideal if HDL or LDL

#fbs - (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)

#restecg - resting electrocardiographic results

#thalach - maximum heart rate achieved

#exang - exercise induced angina (1 = yes; 0 = no)

#oldpeak - ST depression induced by exercise relative to rest

#slope - the slope of the peak exercise ST segment

#ca - number of major vessels (0-3) colored by flourosopy

#thal - 3 = normal; 6 = fixed defect; 7 = reversable defect

#target - have disease or not (1=yes, 0=no)
```

```python
df = df.drop_duplicates()
```

```python
df[(df['age'] < 0) | (df['age'] > 100)].sum()
```

```python
df['age'].value_counts()
```

```python
df['sex'] = df['sex'].map({1:"male", 0:"female"})
```

```python
df.head()
```

```python
df['thal'].value_counts()
```

```python
df = df[df['thal'] > 0]
#thal is thalassemia and  value is 3 6 7 but here we have to remove only 0 due to not enough rows
```

```python
df.shape
```

```python
df['ca'].value_counts()
#ca is number of major vessels affected range is 0 to 3
#0 means no major
#3 means majority
```

```python
#okay so remove 4
df = df[df['ca'] < 4]
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['cp'].value_counts()
```

```python
df['cp'] = df['cp'].map({0:'typical angina',1:'atypical angina',
                            2:'non anginal pain',3:'asymptomatic'})
```

```python
df.head()
```

```python
df['slope'].value_counts()
```

```python
df['slope'] = df['slope'].map({2 : "downsloping", 0 : "upsloping", 1 : "flat"})
```

```python
df.head()
```

```python
df['thal'].value_counts()
```

```python
df['thal'] = df['thal'].map({2:"fixed defect", 3 : "reversible defect", 1 : "normal"})
```

```python
df.head()
```

```python
#trestbps means resting blood pressure measured when a patient is admitted to the hospital
#and range 94 to 200
df[(df['trestbps'] < 94) | df['trestbps'] > 200].sum()
```
