### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
data['MinMaxAge'] = ((data['Age'] - data['Age'].min()) / (data['Age'].max() - data['Age'].min()))
```

```python
data[['Age','MinMaxAge']]
```

```python
data['DecimalScaledAge'] = data['Age'] / 10**(len(str(int(data['Age'].max()))))
data[['Age','DecimalScaledAge']]
```

```python
data['Z-Score Age'] = ((data['Age'] - data['Age'].mean())/data['Age'].std())
data[['Age','Z-Score Age']]
```
