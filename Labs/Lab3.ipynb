### 1)  First, you need to read the titanic dataset from local disk and display first five records



```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("titanic.csv")
data.head(5)
```

### 2)  Identify Nominal, Ordinal, Binary and Numeric attributes from data sets and display all values.



```python
# PassengerId	- Numeric
print(data.Survived.nunique())
print(data.Survived.unique()) #Binary
print(data.Pclass.nunique())
print(data.Pclass.unique()) #ordinal
print(data.Name.nunique()) #Nominal
print(data.Sex.unique()) #Binary
print(data.SibSp.nunique()) #Numeric
print(data.Age.nunique()) #Numeric
print(data.Parch.unique()) #Numeric
print(data.Fare.nunique()) #Numeric
print(data.Ticket.unique()) #Nominal
print(data.Cabin.nunique())
print(data.Cabin.unique()) #Nominal
print(data.Embarked.nunique()) #Nominal
```

### 3)  Identify symmetric and asymmetric binary attributes from data sets and display all values.



```python
print(data.Survived.nunique())
print(data.Survived.unique())#Binary - Asymmetric 
print(data.Sex.nunique()) 
print(data.Sex.unique()) #Binary = Symmetric
```

### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
data.head()
```

```python
values = ["PassengerId","Survived","Pclass","Age","SibSp","Parch","Fare"]
for i in values:
    print(f"{i} : ")
    print("\t Mean = ",data[i].mean())
    print("\t Standard Deviation = ",data[i].std())
    print("\t Mode : ",data[i].mode()[0])
    print(f"\t Min {data[i].min()}")
    print(f"\t Max {data[i].max()}")
    print("\t range : ",data[i].max() - data[i].min())
```

### 8)  For multivariate statistics, you can compute the covariance and correlation between pairs of attributes.



```python
data.corr(numeric_only=True) #between -1 to 1 
#if 1 value increases than other also then gives +ve
#if 1 value increases than other decreases then gives -ve
```

```python
data.cov(numeric_only=True) #difference from mean 
#not fixed that values will be between from -1 to 1
```
